# **Web Scrapping of ABP-NEWS**

```python

chosen_website = "https://www.abplive.com//"
print(f"The chosen website for this scraping task is: {chosen_website}")
```

```python
import requests
import urllib.request
blogurl = "https://www.abplive.com/"
response = requests.get(blogurl)
html=response.text
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(html,"html")
```

```python
headlines = soup.find_all('a')
for h in headlines:
  text = h.text.strip()
  link = h.get("href")
  print(f"{text} : {link}")
```

```python
print(soup.prettify()[:1000])
```

```python
import csv

filename = 'headlines.csv'
with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['text', 'links']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for headline in headlines:
        text = headline.text.strip()
        link = headline.get("href")
        writer.writerow({'text': text, 'links': link})

print(f"Data successfully written to {filename}")
```
