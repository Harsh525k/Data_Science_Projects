```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
iris= pd.read_csv('https://raw.githubusercontent.com/lovnishverma/helloworldirismachinelearning/refs/heads/master/iris.csv')
iris
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
#Encoding of categorical data
encoder=LabelEncoder()
iris['Iris-setosa']=encoder.fit_transform(iris['Iris-setosa'])
iris
```

```python
#split the data as test and train
x=iris.drop('Iris-setosa',axis=1)
y=iris['Iris-setosa']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#Logistic model and its accuracy
model=LogisticRegression()
model.fit(x_train,y_train)
pred= model.predict(x_test)
metrics.accuracy_score(y_test,pred)
print(metrics.accuracy_score(y_test,pred))
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
pred= model.predict(x_test)
metrics.accuracy_score(y_test,pred)
print(metrics.accuracy_score(y_test,pred))
```

# Project
Create an interactive interface in Google Colab using ipywidgets to predict the iris species based on sepal length, sepal width, petal length, and petal width, using a pre-trained machine learning model.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import Button, Output, VBox, FloatSlider, Label


def predict_species(sepal_length, sepal_width, petal_length, petal_width):
  input_data = np.array([[sepal_length, sepal_width, petal_length, petal_width]])
  prediction = model.predict(input_data)
  return prediction[0]

output_widget = Output()
predict_button = Button(description='Predict')

species_map = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}

# Define the input widgets
sepal_length_widget = FloatSlider(min=0, max=10, step=0.1, description='Sepal Length:')
sepal_width_widget = FloatSlider(min=0, max=10, step=0.1, description='Sepal Width:')
petal_length_widget = FloatSlider(min=0, max=10, step=0.1, description='Petal Length:')
petal_width_widget = FloatSlider(min=0, max=10, step=0.1, description='Petal Width:')


def on_predict_button_clicked(b):
  with output_widget:
    output_widget.clear_output()
    sepal_length = sepal_length_widget.value
    sepal_width = sepal_width_widget.value
    petal_length = petal_length_widget.value
    petal_width = petal_width_widget.value

    prediction = predict_species(sepal_length, sepal_width, petal_length, petal_width)
    species_name = species_map.get(prediction, 'Unknown Species')
    print(f"Predicted Species: {species_name}")

predict_button.on_click(on_predict_button_clicked)

display(VBox([sepal_length_widget, sepal_width_widget, petal_length_widget, petal_width_widget, predict_button, output_widget]))



# # Test with values for Iris-setosa
# sepal_length_widget.value = 5.1
# sepal_width_widget.value = 3.5
# petal_length_widget.value = 1.4
# petal_width_widget.value = 0.2
# predict_button.click()

# # Test with values for Iris-versicolor
# sepal_length_widget.value = 6.0
# sepal_width_widget.value = 2.9
# petal_length_widget.value = 4.5
# petal_width_widget.value = 1.5
# predict_button.click()

# # Test with values for Iris-virginica
# sepal_length_widget.value = 6.7
# sepal_width_widget.value = 3.0
# petal_length_widget.value = 5.2
# petal_width_widget.value = 2.3
# predict_button.click()
```
